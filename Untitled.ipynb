```python
import pandas as pd
import glob, os   
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
columns = ['Model', 'Device', 'Threads'] 
```

```python
def averages (folder='hikeypower'):
    return pd.DataFrame(map(lambda x:Path(x).stem.split('-')+[pd.read_csv(x)['Main Avg Current (mA)'].mean()], glob.glob(os.path.join('', folder+'/*'))), columns=columns+['mA']
                    ).astype({'Threads': 'int64', }).sort_values(columns, ascending=[False, True, True]).reset_index(drop=True)

def plotByThreads (df, ylabel='ms/frame'): #draws all models (Model=quant..., Device=CPU, GPU)
    fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(30,5), sharey=True)
    for (key, ax) in zip(df.groupby(['Model', 'Device']).groups.keys(), axes.flatten()):
        df.groupby(['Model', 'Device']).get_group(key).plot(ax=ax, x='Threads', legend=True, title=' '.join(key))
        ax.yaxis.grid(True)
        ax.set_ylabel(ylabel, color='black')
```

# Power measurements

## Default scaling governor (schedutil) & frequencies

```python
power = pd.merge(averages (folder='hikeypower'), pd.read_csv("Testi.csv"), on=['Model','Device','Threads'])
power
```

## Userspace scaling governor and minimum frequencies (little CPU = 509MHz, big CPU = 682)

```python
powermin = pd.merge(averages (folder='hikeyMinFreq'), pd.read_csv("TestiMinFreq.csv"), on=['Model','Device','Threads'])
powermin
```

## Hikey with cooling - always MAX frequency

```python
powermax = pd.merge(averages (folder='hikeyMaxFreq'), pd.read_csv("TestiMaxFreq.csv"), on=['Model','Device','Threads'])
powermax
```

```python
def groupedGraphs(df):
    grouped = df.groupby(['Model', 'Device'])
    columns = list(grouped.obj.columns)
    
    columns = list(filter(lambda x: x not in  ['Model', 'Device','Lfreq', 'Bfreq', 'Threads'] , columns))
    
    counter = int('1'+ str(len(columns))+'1')
    
    plt.figure(figsize=(len(columns)*5, 5))
    
    for col in columns:
        plt.subplot(counter)
        for key in grouped.groups.keys():
                plt.plot(grouped.get_group(key)['Threads'], grouped.get_group(key)[col], label=' '.join(key))

        plt.legend()
        plt.ylabel(col)
        plt.xlabel('Threads')
        counter += 1  
```
