```python
import pandas as pd
import glob, os   
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
columns = ['Model', 'Device', 'Threads'] 
```

# Measurements from TF

```python
time = pd.read_csv("Testi.csv")
time.head()
```

# Power measurements

```python
power = pd.DataFrame(map(lambda x:Path(x).stem.split('-')+[pd.read_csv(x)['Main Avg Current (mA)'].mean()], glob.glob(os.path.join('', "hikeypower/*"))), columns=columns+['mA']
                    ).astype({'Threads': 'int64'}).sort_values(columns, ascending=[False, True, True]).reset_index(drop=True)
power
```

# Merge by keys: Model, Device, Threads

```python
merged = pd.merge(time, power, on=['Model','Device','Threads'])
merged.head()
```

# Plot merged data

```python
grouped = time.groupby(['Model', 'Device'])

ncols=2
nrows = int(np.ceil(grouped.ngroups/ncols))

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(30,5), sharey=True)


for (key, ax) in zip(grouped.groups.keys(), axes.flatten()):
    grouped.get_group(key).plot(ax=ax, x='Threads', legend=True, title=' '.join(key))
    ax.yaxis.grid(True)
    ax.set_ylabel('ms/frame', color='black')
    
#     ax.yaxis.set_tick_params(labelbottom=True)

    
    ax2 = ax.twinx()
    power.groupby(['Model', 'Device']).get_group(key).sort_values(by=['Threads']).plot(ax=ax2, x='Threads', color='r', legend=False)
    ax2.set_ylabel('mA', color='r')



ax.legend()
plt.show()

# for key, item in grouped:
#     grouped.get_group(key).sort_values(by=['Threads']).plot(x='Threads', legend=True, title=' '.join(key))
```
